## Расчёты и добавление результатов в таблицу

### Расчет цены квадратного метра

```python
#рассчитаем стоимость квадратного метра жилья с помощью функции

def coast_metr(row):
    
    total_area = row['total_area']
    last_price = row['last_price']   
    
    coast_za_metr = last_price/total_area
    
    return coast_za_metr
    
data['coast_metr'] = data.apply(coast_metr, axis=1)
```

### Расчет дня недели, месяца и года публикации объявления


```python
#рассчитаем необходимые параметры с помощью атрибутов класса DatetimeIndex 

data['day_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).day_name()

data['month_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).month 

data['year_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).year 
```

```python
#заменим названия дней недели 
data['day_exposition'] = data['day_exposition'].\
                    replace(to_replace=['Monday','Tuesday', 'Wednesday', 
                                        'Thursday',  'Friday', 'Saturday', 'Sunday'],
                    value=['Понедельник', 'Вторник', 'Среда', 'Чертверг', 
                           'Пятница', 'Суббота', 'Воскресенье'])
```

### Расчет этажа квартиры


```python
#применим функцию apply для расчета этажа квартиры

def floor_kvat(row):
    
    floors_total = row['floors_total']
    floor = row['floor']
    
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'
    
data['cat_floor'] = data.apply(floor_kvat, axis=1)    
```

### Расчет соотношения жилой и общей площади, а также отношения площади кухни к общей

```python
#применим функцию apply для расчета соотношения жилой и общей площади, а также отношения площади кухни к общей

def liv_total_area(row):
    
    total_area = row['total_area']
    living_area = row['living_area']
    
    liv_total = living_area/total_area
    
    return liv_total

def kitch_total_area(row):
    
    total_area = row['total_area']
    kitchen_area = row['kitchen_area']
    
    kitchen_total = kitchen_area/total_area
    
    return kitchen_total
    
data['liv_total_area'] = data.apply(liv_total_area, axis=1)  

data['kitch_total_area'] = data.apply(kitch_total_area, axis=1)    
```

### Вывод

В данном разделе были применены методы по работе с датами для получения дня, месяца и года размещения объявления, также был примененен метод apply для расчета дополнительных параметров по характеристикам квартир.

```python
#поделим все расстояния до центра, представленные в нашем датафрейме, на условные группы

def city_center(row):
    
    cityCenters_nearest=row['cityCenters_nearest']
        
    if cityCenters_nearest <= 10000:
        return 'до 10 км'
    if 10000 < cityCenters_nearest <= 20000:
        return 'от 10 до 20 км'
    if 20000 < cityCenters_nearest <= 40000:
        return 'от 20 до 40 км'
    return 'более 40 км'
    
data['cityCenters_nearest_cat'] = data.apply(city_center, axis=1)
```

```python
#построим сводную на основе данных о цене квартир и категории по удаленности от центра

price_cityCenters_nearest = data.pivot_table(index='cityCenters_nearest_cat', values='last_price', aggfunc='median')

price_cityCenters_nearest.sort_values(by='last_price', ascending=False)
```

<div class="alert alert-info">
<b>Комментарий студента:</b>
<br> 
По диаграмме рассеяния невозможно сказать есть ли зависимость между удаленностью от центра и стоимостью квартир. Проведем анализ с помощью сводной таблицы.
<br>

```python
#поделим все расстояния до центра, представленные в нашем датафрейме, на условные группы

def city_center_cen(row):
    
    cityCenters_nearest_km = row['cityCenters_nearest_km']
        
    if cityCenters_nearest_km <= 1:
        return 'до 1 км'
    if 1 < cityCenters_nearest_km <= 2:
        return 'от 1 до 2 км'
    return 'более 2 км'
    
data_center['cityCenters_nearest_cat_km'] = data_center.apply(city_center_cen, axis=1)
data_spb['cityCenters_nearest_cat_km'] = data_spb.apply(city_center_cen, axis=1)
```

```python
#построим сводную на основе данных о цене квартир и категории по удаленности от центра

price_cityCenters_nearest_center = data_center.pivot_table(index='cityCenters_nearest_cat_km', values='last_price', aggfunc='median')
price_cityCenters_nearest_spb = data_spb.pivot_table(index='cityCenters_nearest_cat_km', values='last_price', aggfunc='median')
```

```python
#объединим таблицы для просмотра результатов

price_cityCenters_nearest_center_spb = price_cityCenters_nearest_center.merge(price_cityCenters_nearest_spb, on='cityCenters_nearest_cat_km').sort_values(by='last_price_x')

price_cityCenters_nearest_center_spb
```

```python
price_cityCenters_nearest_center_spb.columns = ['цена кв в центре', 'цена кв по городу']
```

```python
price_cityCenters_nearest_center_spb.plot(grid=True)
```

<div class="alert alert-info">
<b>Комментарий студента:</b>
<br> 
Существует зависимость между расстоянием до центра города и стоимостью квартиры. Квартиры ближе к центру стоят значительно дороже остальных. Данный тренд наблюдается как у квартир в центре города, так и в целом по городу.

- Зависимость между ценой и площадью квартиры есть, хотя она и не очень сильная. При этом, она положительная, то есть при увеличении площади квартиры увеличивается ее стоимость.
____________________________________
- Существует прямая зависимость между стоимостью квартиры и количеством комнат, но это работает только если комнат в квартире не более пяти, если количество комнат больше, данный тренд не соблюдается. Коэффициент корреляции Пирсона также показал слабую зависимость данных параметров.
_________________________________________
- Существует зависимость между расстоянием до центра города и стоимостью квартиры. Квартиры ближе к центру стоят значительно дороже остальных, а квартиры вдали от центра стоят почти в три раза дешевле тех, что в центре.
___________________________________
- Существует зависимость между тем на каком этаже находится квартира и ее стоимостью. Как показывают данные из сводной таблицы, квартиры на первом этаже стоят значительно дешевле квартир, находящихся на последнем, либо на другом, отлчном от первого, этаже. Квартиры на последнем этаже также стоят дешевле квартир, находящихся на другом этаже, ввиду того, что есть различные риски, например, может протечь крыша.
- По сравнению с квартирами в целом по городу, квартиры на первом этаже в центре стоят значительно дешевле квартир на других этажах.

- Самое дорогое жилье в Санкт-Петербурге, Пушкине и Кудрово, а самое дешевое - в Всеволжске, Гатчине и Выборге.

- Квартиры в центре, объявления которых размещены в зимний период, стоят дороже квартир, объявления которых размещены в другие месяцы.
