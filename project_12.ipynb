<div class="alert alert-info">
<b>Комментарий студента:</b>
<br>
Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Подготовим переменные и формулу для дальнейшего расчета.
</div>

```python
#для начала создадим переменнные

concentrate = gold_recovery_train['rougher.output.concentrate_au']
feed = gold_recovery_train['rougher.input.feed_au']
tail = gold_recovery_train['rougher.output.tail_au']
init_recovery = gold_recovery_train['rougher.output.recovery']                                  
```

```python
#формула для расчета 

calculated_recovery = (concentrate * (feed - tail)) / (feed * (concentrate - tail))* 100 

#далее рассчитаем значение средней абсолютной ошибки

print('Средняя абсолютная ошибка:', mean_absolute_error(init_recovery, calculated_recovery))
```

### Исследование изменения концентраций металлов (Au, Ag, Pb) на различных этапах очистки

<div class="alert alert-info">
<b>Комментарий студента:</b>
<br>
Для начала составим функцию, которая бы нам позволила автоматизировать построение графиков для исследования изменения концентраций металлов на различных этапах очистки.
</div>

```python
def metal_concenrtate(metal, name):
    
    plt.figure(figsize=[17,9])

    plt.hist(gold_recovery_full[f'rougher.input.feed_{metal}'], bins=100, label = 'В исходном сырье', alpha=.5)
    plt.hist(gold_recovery_full[f'rougher.output.concentrate_{metal}'], bins=100, label = 'На этапе флотации', alpha=.5)
    plt.hist(gold_recovery_full[f'primary_cleaner.output.concentrate_{metal}'], bins=100, label = 'На этапе первичной очистки', alpha=.5)
    plt.hist(gold_recovery_full[f'final.output.concentrate_{metal}'], bins=100, label = 'В финальном концентрате', alpha=.5)
    
    plt.xlabel(f'Концентрация {name} на различных этапах очистки')
    plt.ylabel('Количество результатов')

    plt.legend(loc='upper left')

    plt.title(f'Гистограмма распределения концентрации {name} на различных этапах очистки')
    plt.show()
```

```python
def feed_size(stage, name):
    
    plt.figure(figsize=[17,7])

    plt.hist(gold_recovery_train[f'{stage}.feed_size'], bins=100, label = 'На обучающей выборке', alpha=.5)
    plt.hist(gold_recovery_test[f'{stage}.feed_size'], bins=100, label = 'На тестовой выборке', alpha=.5)
    
    plt.xlabel('Размер гранул сырья')
    plt.ylabel('Количество результатов')

    plt.legend(loc='upper right')

    plt.title(f'Распределение размеров гранул {name}')
    plt.show()
```

### Написание функции для вычисления итогового sMAPE
 

<div class="alert alert-info">
<b>Комментарий студента:</b>
<br>
Для начала напишем формулу для sMAPE.
</div>

```python
def sMAPE_score(target, predictions):
    
    sMAPE = (1/len(target) * np.sum(2*np.abs(target - predictions)/(np.abs(target) + np.abs(predictions))) * 100)
    
    return sMAPE
```

<div class="alert alert-info">
<b>Комментарий студента:</b>
<br>
Далее напишем формулу для итогового sMAPE.
</div>

```python
def sMAPE_result(rougher_sMAPE, final_sMAPE):
    
    sMAPE_res = rougher_sMAPE * 0.25 + final_sMAPE * 0.75
    
    return sMAPE_res
```
